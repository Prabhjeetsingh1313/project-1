# import libraries

```python
import pandas as pd
import numpy as np
```

# load dataset_1

```python
dataset_1 = pd.read_csv("dataset_1.csv")
```

```python
dataset_1.head()
```

```python
# read dimension of dataset-1
dataset_1.shape
```

# load dataset_2

```python
dataset_2 = pd.read_excel("dataset_2.xlsx")
dataset_2.head()
```

# merge datasets 1 & 2

```python
bike_count = pd.merge(dataset_1, dataset_2,on='instant',how='inner')
bike_count.head()
```

# drop unnecessary columns

```python
bike_count.drop('instant',axis=1,inplace=True)
bike_count.head()
```

```python
bike_count.drop('yr',axis=1,inplace=True)
bike_count.head()
```

```python
bike_count.drop('hr',axis=1,inplace=True)
bike_count.head()
```

```python
bike_count.drop('mnth',axis=1,inplace=True)
bike_count.head()
```

# check the dimension of the dataset

```python
bike_count.shape
```

# central tendency

```python
bike_count.describe()
```

# check the datatype & summary of dataset


```python
bike_count.info()
```

# treat missing values

```python
# to find the missing numbers
bike_count.isnull().sum()
```

```python
# to fill value with mean (atemp) column
bike_count['atemp'] = bike_count['atemp'].fillna(bike_count['atemp'].mean())
bike_count.isnull().sum()
```

# to rename the columns


```python
bike_count = bike_count.rename(columns={'dteday':'date'})
bike_count.head()
```

```python
#converting date into integars value
bike_count['date']=pd.to_datetime(bike_count['date'],dayfirst=True).dt.strftime('%Y%m%d').astype('int')
bike_count.head()
```

```python
# to check datatypes
bike_count.info()
```

# load dataset_3

```python
dataset_3 = pd.read_excel("dataset_3.xlsx")
dataset_3.head()
```

# drop unnecessary columns

```python
dataset_3.drop('instant',axis=1,inplace=True)
dataset_3.drop('yr',axis=1,inplace=True)
dataset_3.drop('mnth',axis=1,inplace=True)
dataset_3.drop('hr',axis=1,inplace=True)
dataset_3.head()
```

# check data types

```python
dataset_3.info()
```

# rename date column and chage date into integer value

```python
dataset_3 = dataset_3.rename(columns={'dteday':'date'})
dataset_3['date']=pd.to_datetime(dataset_3['date'],dayfirst=True).dt.strftime('%Y%m%d').astype('int')
dataset_3.head()
```

```python
# check data types
dataset_3.info()
```

```python
# check the missing values
dataset_3.isnull().sum()
```

```python
# dropping unnecessary columns
bike_count.drop('Unnamed: 0',axis=1,inplace=True)
bike_count.head()
```
